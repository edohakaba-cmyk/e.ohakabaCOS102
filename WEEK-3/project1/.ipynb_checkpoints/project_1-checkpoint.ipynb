```python

import tkinter as tk
from tkinter import messagebox
import pandas as pd


# 1. Load the dataset
# Make sure "GIG-logistics(in).csv" is in the same folder as this script
try:
    rf = pd.read_csv("GIG-logistics(in).csv")
except FileNotFoundError:
    # This creates a temporary dataframe so you can still test the UI
    data = {
        'Name': ['Mary', 'John', 'Alice'],
        'Department': ['Logistics', 'Logistics', 'Accounting']
    }
    rf = pd.DataFrame(data)
    print("System: CSV not found. Loading internal test data...")

def welcomeMessage(username):
    # Create a success popup window
    window = tk.Toplevel(root)
    window.title("Admin Box")
    window.geometry("350x250")

    tk.Label(window, text=f"Welcome {username}", font=("Arial", 20, "bold")).pack(pady=10)
    tk.Label(window, text="This is Python GUI with Tkinter",font = ("Arial",15)).pack(pady=5)
    tk.Label(window, text ="Have a nice day",font = ("Arial",15)).pack(pady=5)
    tk.Label(window, text="🤗",font = ("Arial",35)).pack(pady=5)

def submit():
    # Get values from the entry boxes
    name = name_entry.get().strip()
    department = department_entry.get().strip()

    # Basic Validation
    if name == "" or department == "":
        messagebox.showwarning("Input Error", "Please enter both name and department.")
        return

    # 2. Search Logic
    # We use .str.lower() to make the search case-insensitive
    employee_match = rf[(rf['Name'].str.lower() == name.lower()) & 
                        (rf['Department'].str.lower() == department.lower())]

    if not employee_match.empty:
        # 3. Find other members in the same department
        all_members = rf[rf['Department'].str.lower() == department.lower()]['Name'].tolist()
        
        # Remove the current user from the list of "other members"
        # We find the exact spelling from the CSV to ensure .remove() works
        current_user_exact_name = employee_match.iloc[0]['Name']
        if current_user_exact_name in all_members:
            all_members.remove(current_user_exact_name)
        
        member_list = ", ".join(all_members) if all_members else "No other members"
        
        # Show Info Box
        messagebox.showinfo("Welcome", 
                            f"Welcome, {current_user_exact_name}!!\n\n"
                            f"Other members in {department}:\n{member_list}")
        
        # Open the additional welcome window
        welcomeMessage(current_user_exact_name)
    else:
        # Polite refusal
        messagebox.showwarning("Access Denied", "Employee does not exist in our records.")

# 4. Main GUI Setup
root = tk.Tk()
root.title("GIG Logistics - Staff Login")
root.geometry("400x300")

# Header
tk.Label(root, text="GIG ", font=("Arial", 25, "bold"), fg="purple").pack(pady=0)
tk.Label(root, text = "Logistics", font = ('Arial',35),fg = "Purple").pack(pady = 1)

tk.Label(root, text = "The ones you can trust...😎", font = ('Arial',10),fg = "purple").pack(pady = 1)
# Username Section
tk.Label(root, text="Enter Name:").pack()
name_entry = tk.Entry(root, width=30)
name_entry.pack(pady=3)

# Department Section
tk.Label(root, text="Enter Department:").pack()
department_entry = tk.Entry(root, width=30)
department_entry.pack(pady=3)

# Submit Button
submit_button = tk.Button(root, text="Submit", command=submit, bg="Black", fg="white", width=17)
submit_button.pack(pady=5)



# Run the application
root.mainloop()
```
