```python
import tkinter as tk
from tkinter import messagebox


class DeliveryService:
    def __init__(self):
        # Pricing rules
        self.rules = {
            "PAU": {">=10": 2000, "<10": 1500},
            "Epe": {">=10": 5000, "<10": 4000}
        }

    def calculate_cost(self, location, weight):
        location = location.strip().title()

        if location not in self.rules:
            return None

        if weight >= 10:
            return self.rules[location][">=10"]
        else:
            return self.rules[location]["<10"]


class DeliveryGUI:
    def __init__(self, root):
        self.service = DeliveryService()
        self.root = root
        self.root.title("Delivery Cost Calculator")
        self.root.geometry("400x300")

        # Title
        self.title_label = tk.Label(root, text="Delivery Price Checker", font=("Arial", 14, "bold"))
        self.title_label.pack(pady=10)

        # Location
        self.loc_label = tk.Label(root, text="Enter Location (PAU or Epe):")
        self.loc_label.pack()
        self.loc_entry = tk.Entry(root)
        self.loc_entry.pack(pady=5)

        # Weight
        self.weight_label = tk.Label(root, text="Enter Package Weight (kg):")
        self.weight_label.pack()
        self.weight_entry = tk.Entry(root)
        self.weight_entry.pack(pady=5)

        # Button
        self.calc_btn = tk.Button(root, text="Calculate", command=self.calculate)
        self.calc_btn.pack(pady=10)

        # Result
        self.result_label = tk.Label(root, text="", font=("Arial", 12))
        self.result_label.pack(pady=15)

    def calculate(self):
        location = self.loc_entry.get()
        weight_input = self.weight_entry.get()

        try:
            weight = float(weight_input)
        except ValueError:
            messagebox.showerror("Invalid Input", "Please enter a valid number for weight.")
            return

        cost = self.service.calculate_cost(location, weight)

        if cost is None:
            messagebox.showerror("Invalid Location", "We only deliver to PAU or Epe.")
            self.result_label.config(text="")
        else:
            self.result_label.config(text=f"Delivery Cost: ₦{cost}")


if __name__ == "__main__":
    root = tk.Tk()
    app = DeliveryGUI(root)
    root.mainloop()
```
